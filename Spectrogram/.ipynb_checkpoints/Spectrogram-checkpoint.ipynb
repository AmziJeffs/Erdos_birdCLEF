```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os
import IPython.display as ipd
```

```python
class CONFIG:
    # Mel Spec:
    SHAPE = (256, 256)
    FRAME_SIZE = 2048 # n_fft
    AUDIO_DURATION = 5 # 5 sec
    N_MELS = SHAPE[0]
    MIN_SAMPLES = 100
    
    # Directory:
    INPUT_DIR = "../data/"
    METADATA = os.path.join(INPUT_DIR, "train_metadata.csv")
    TRAIN_AUDIO = os.path.join(INPUT_DIR, "train_audio")
```

```python
data = pd.read_csv(CONFIG.METADATA)
data.head(4)
```

```python
#Modify this based on where you've stored your audio files
sample_path = '../data/train_audio/'+data.iloc[100]['filename']
```

```python
sample_path
```

```python
signal, sr = librosa.load(sample_path)
print('Signal:', signal)
print('Shape of signal:', np.shape(signal))
print('Sampling Rate (KHz):', sr)
print('Duration of audio:', np.shape(signal)[0]/sr)
plt.plot(signal)
plt.show()
ipd.Audio(sample_path)
```

```python
D = np.abs(librosa.stft(signal, n_fft=CONFIG.FRAME_SIZE, hop_length=512))
plt.plot(D,  color='blue')
plt.show()
```
